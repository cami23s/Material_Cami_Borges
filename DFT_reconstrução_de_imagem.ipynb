```python
import numpy as np
import matplotlib.pyplot as plt

# Função para reconstruir a imagem a partir da DFT
def reconstruct_image(F):
    F_shifted = np.fft.ifftshift(F)  # Deslocamento da DFT para o centro
    f_reconstructed = np.fft.ifft2(F_shifted)  # Transformada inversa da DFT
    f_reconstructed = np.abs(f_reconstructed)  # Valor absoluto para obter a imagem reconstruída
    return f_reconstructed

# Carregar a imagem
image = plt.imread('C:\\Users\\camil\\Downloads\\Bonten.png')

# Aplicar a DFT na imagem
F = np.fft.fft2(image)

# Processamento na DFT (remoção de frequências altas)
# Aqui vamos zerar as frequências altas em um quadrante da DFT
rows = len(image)
cols = len(image[0])
F[rows//2:, cols//2:] = 0

# Reconstruir a imagem
image_reconstructed = reconstruct_image(F)

# Exibir a imagem original e a imagem reconstruída
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(image_reconstructed, cmap='gray')
plt.title('Reconstruída')

plt.show()
```
