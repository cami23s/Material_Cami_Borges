```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
M_solar = 1.989e30  # Massa do sol em kg
mu_e = 2.0  # Número médio de partículas por elétron (assumido para carbono-oxigênio)
M_Ch_base = 1.44  # Limite de Chandrasekhar em massas solares

# Função para calcular o limite de Chandrasekhar
def limite_chandrasekhar(mu_e):
    return M_Ch_base * (2 / mu_e)**2

# Parâmetros de Sirius B
M_sirius_B = 1.018 * M_solar  # Massa de Sirius B (aproximadamente 1.018 massas solares)

# Gerar variação de valores de mu_e
mu_e_values = np.linspace(1.5, 3.0, 100)
mass_limits = [limite_chandrasekhar(mu) for mu in mu_e_values]

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(mu_e_values, mass_limits, label='Limite de Chandrasekhar', color='b')
plt.axhline(y=M_sirius_B/M_solar, color='r', linestyle='--', label='Massa de Sirius B')

plt.title('Limite de Chandrasekhar vs. Massa de Sirius B')
plt.xlabel('Número médio de partículas por elétron, $\mu_e$')
plt.ylabel('Limite de Chandrasekhar (Massas Solares)')
plt.legend()

plt.grid(True)
plt.show()
```
