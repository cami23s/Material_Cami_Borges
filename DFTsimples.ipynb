```python
import numpy as np
import matplotlib.pyplot as plt

f = np.zeros((30, 30))
f[4:24, 12:17] = 1
plt.imshow(f, cmap='gray')
plt.show()

F = np.fft.fft2(f)
F2 = np.log(np.abs(F))
plt.imshow(F2, vmin=-14, vmax=24, cmap='jet')
plt.colorbar()
plt.show()

F2 = np.fft.fftshift(F)
F3 = np.log(np.abs(F2))
plt.imshow(F3, vmin=-1, vmax=5, cmap='jet')
plt.colorbar()
plt.show()
```
